### 字段差异对比

表格字段：页面中表格的头部字段；大部分功能由纯表格组成。

文本字段：页面中非表格字段,大部分有文本字段的页面文本字段数量都少于12个。

功能内页中的字段数目有很大差异，最长的有49个字段（4个tab页),比如，上海LOF基金->LOF拆分.

```python
print('表格字段较多的页面：')
gc=g.loc[:,['表格字段','文本字段']].applymap(lambda x:str(x).count('|')+1)
df=pd.concat([g,gc],axis=1)
df.columns=['大类', '功能', '子功能', '表格字段', '文本字段', '编号', '软件(东财0|同花顺1)', '表格字段数',
       '文本字段数']
df.sort_values('表格字段数',ascending=False).head(15)
```

```python
df1=df.loc[:,['功能','表格字段数']]
df1['二级功能']=df['功能']+'->'+df['子功能']
df1.set_index('二级功能',inplace=True)
df1.sort_values('表格字段数',ascending=False,inplace=True)
df1.head(20).plot(kind='bar',figsize=(16,5))
```

### 字段差异
东方财富：表格字段差异对照

```python
def diff(column=3):
    for i in range(1,103):
        try:
            dc=set(g[(g['软件(东财0|同花顺1)']==0) & (g['编号']==i)].iloc[0,column].split('|')) 
            ths=set(g[(g['软件(东财0|同花顺1)']==1) & (g['编号']==i)].iloc[0,column].split('|')) 
            print('第%s项功能'%i)
            print('东财：【%s】'%(g[(g['软件(东财0|同花顺1)']==0) & (g['编号']==i)].iloc[0,1]+\
                       '->'+g[(g['软件(东财0|同花顺1)']==0) & (g['编号']==i)].iloc[0,2])+'  VS  '+\
              '同花顺：【%s】'%(g[(g['软件(东财0|同花顺1)']==1) & (g['编号']==i)].iloc[0,1]+\
                        '->'+g[(g['软件(东财0|同花顺1)']==1) & (g['编号']==i)].iloc[0,2]))
    
            print('共同字段:')
            print(dc & ths)
            print('东财特有字段:')
            print(dc - ths)
            print('同花顺特有字段:')
            print(ths - dc)
            print('-'*100)
        except:
            print('第%s项功能为不重合{东财：0，同花顺：1}'%i)
            print(g[g.编号==i].iloc[0,-1])
#表格字段差异
#diff(3)



#文本字段差异
diff(4)
```
