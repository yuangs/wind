```python
def shf_oi(start=20170703,end=20170705):
    bdays=pd.bdate_range(start=str(start),end=str(end)).strftime('%Y%m%d').tolist()
    data=pd.DataFrame()
    for day in bdays:
        url="http://www.shfe.com.cn/data/dailydata/kx/pm%s.dat"%day
        try:
            text=json.loads(requests.get(url).text)
            if text:
                df=pd.DataFrame(text['o_cursor'])
                df['DATE']=text['report_date']
                data=data.append(df,ignore_index=True)
                print(day)
        except:
            print('%s no data!')
    data.to_csv(r"C:\Users\Administrator\Desktop\shf_oi.csv")
#         return data
#         data=df.append(data,ignore_index=True)
```

```python
def monthly_oi(company='永安期货',code='rb'):
    import sqlite3
    con=sqlite3.connect(r'd:\aliyun\www\my_flask\marketPrice\shfoi.db')
    Lsql=r"select substr(date,1,6) as month, name,longcom,sum(long) as total_long FROM 'shfoi.db' \
    WHERE longcom like'%{company}%' and code like '%{code}%' GROUP BY longcom,substr(date,1,6) \
    ORDER BY date desc;".format(company=company,code=code)
    Ssql=r"select substr(date,1,6) as month, name,shortcom,sum(short) as total_short FROM 'shfoi.db' \
    WHERE shortcom like'%{company}%' and code like '%{code}%' GROUP BY shortcom,substr(date,1,6) \
    ORDER BY date desc".format(company=company,code=code)
#     print(Lsql,'\n',Ssql)
    df1=pd.read_sql(Lsql,con)
    df2=pd.read_sql(Ssql,con)
    df2.sort_index(inplace=True)
    df1.sort_index(inplace=True)
    df1['total_short']=df2.total_short
    df1.columns=['month','name','com','long','short']
    df1['ratio']=df1.long/df1.short
    oi=df1.sort_values('month',ascending=False)
    oi.set_index('month',inplace=True)
    return oi
```

```python
def dce_oi(code='j',start=20170803,end=20170805):
    bdays=pd.bdate_range(start=str(start),end=str(end)).strftime('%Y%m%d').tolist()
    data=pd.DataFrame()
    for day in bdays:
        url="http://www.dce.com.cn/publicweb/quotesdata/memberDealPosiQuotes.html?memberDealPosiQuotes.variety={code}&memberDealPosiQuotes.trade_type=0&year={year}&month={month}&day={day}&contract.contract_id=all&contract.variety_id={code}".format(code=code,year=day[:4],month=int(day[4:6])-1,day=day[6:])
        try:
            df=pd.read_html(url)[1][1:-1]
            df['DATE']=day
            data=data.append(df,ignore_index=True)
            print(day)
        except:
            print('%s no data!'%day)
    data.to_csv(r"dce_oi_%s_%s.csv"%(day[:4],code))
#         return data
#         data=df.append(data,ignore_index=True)
```
